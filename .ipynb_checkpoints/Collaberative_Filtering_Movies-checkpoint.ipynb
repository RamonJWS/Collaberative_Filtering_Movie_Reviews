```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
class CollaberativeFiltering():
    
    def __init__(self, file_path):
        
        df = pd.read_csv("data/ratings_small.csv")
        user_movies = df.pivot(index="userId", columns="movieId", values="rating")
        user_movies.reset_index(inplace=True)
        user_movies.columns = [str(x) for x in user_movies.columns]
        user_movies.fillna(0, inplace=True)
        user_movies_np = user_movies.iloc[:,1:].to_numpy()
        
        self.user_movies = user_movies_np
        
    def data_split(self, split):
        
        masked_ratings = np.zeros(self.user_movies.shape)

        for i in range(masked_ratings.shape[0]):
            for j in range(masked_ratings.shape[1]):
                if self.user_movies[i,j] != 0:
                    masked_ratings[i,j] = np.random.choice([0, self.user_movies[i,j]], size=1, p=[split,1-split])[0]
                    
        self.masked_ratings = masked_ratings
        self.true_ratings = self.user_movies - self.masked_ratings
        
    def user_user_similarity(self):
        
        user_user_dot = np.matmul(self.masked_ratings, self.masked_ratings.T)
        user_eclid = np.linalg.norm(self.masked_ratings, axis=1)
        
        user_user_sim = np.zeros((len(user_eclid),len(user_eclid)))

        for i in range(len(user_eclid)):
            for j in range(len(user_eclid)):
                user_user_sim[i][j] = user_user_dot[i][j] / (user_eclid[i] * user_eclid[j])
                
        self.user_user_sim = user_user_sim
        
    def predict(self):
        
        predictions = np.zeros(self.masked_ratings.shape)

        for i in range(predictions.shape[0]):
            for j in range(predictions.shape[1]):
                if self.masked_ratings[i][j] == 0:
                    sum_of_sim = np.sum(self.user_user_sim[i,:][self.masked_ratings[:,j] > 0])
                    if sum_of_sim == 0:
                        predictions[i][j] = np.mean(self.user_movies[:,j][self.user_movies[:,j] != 0])
                    else:
                        predictions[i][j] = np.dot(self.masked_ratings[:,j], self.user_user_sim[i,:]) / sum_of_sim
                        
        self.predictions = predictions
        
    def evaluation(self):
        
        pred = self.predictions[self.true_ratings != 0]
        pred_no_nan = np.round(np.nan_to_num(pred, nan=0) * 2) / 2
        
        true = self.true_ratings[self.true_ratings != 0]
        rmse = np.sqrt(((pred_no_nan - true) ** 2).mean())
        
        self.predicted_ratings_only = pred_no_nan
        self.true_ratings_only = true
        self.rmse = rmse
        
    def plot(self):
        
        plt.figure(figsize=(12,5))
        plt.title("error between predicted and actual ratings")
        plt.xlabel("Rating Error")
        sns.histplot((self.predicted_ratings_only - self.true_ratings_only));
        plt.xticks(np.arange(-5, 5.5, 0.5).tolist());
```

```python
test = CollaberativeFiltering("data/ratings_small.csv")
test.data_split(0.1)
test.user_user_similarity()
test.predict()
test.evaluation()
test.plot()
```
